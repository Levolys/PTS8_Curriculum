#Conversion

```
filenames = ("shapeset1_1c_2s_3po.5000.test.amat",
         "shapeset1_1cs_2p_3o.5000.test.amat",
         "shapeset1_1csp_2o_3.5000.test.amat",
         "shapeset1_1cspo_2_3.5000.test.amat",
         "shapeset2_1c_2s_3po.5000.test.amat",
         "shapeset2_1cs_2p_3o.5000.test.amat",
         "shapeset2_1csp_2o_3.5000.test.amat",
         "shapeset2_1cspo_2_3.5000.test.amat",
         "shapeset1_1c_2s_3po.10000.train.amat",
         "shapeset1_1cs_2p_3o.10000.train.amat",
         "shapeset1_1csp_2o_3.10000.train.amat",
         "shapeset1_1cspo_2_3.10000.train.amat",
         "shapeset2_1c_2s_3po.10000.train.amat",
         "shapeset2_1cs_2p_3o.10000.train.amat",
         "shapeset2_1csp_2o_3.10000.train.amat",
         "shapeset2_1cspo_2_3.10000.train.amat",
         "shapeset1_1c_2s_3po.5000.valid.amat",
         "shapeset1_1cs_2p_3o.5000.valid.amat",
         "shapeset1_1csp_2o_3.5000.valid.amat",
         "shapeset1_1cspo_2_3.5000.valid.amat",
         "shapeset2_1c_2s_3po.5000.valid.amat",
         "shapeset2_1cs_2p_3o.5000.valid.amat",
         "shapeset2_1csp_2o_3.5000.valid.amat",
         "shapeset2_1cspo_2_3.5000.valid.amat")


for f in filenames :
  
    with open(re.sub(r'amat',r'csv',f), 'wb') as csvfile:
        csvwriter = csv.writer(csvfile, dialect='excel')
        csvwriter.writerow(["pixel"+str(ind) for ind in range(1024)]+["shape","color","x_centre","y_centre","rotation","size","elongation"])

        with open(f, 'rb') as amatfile :
        
            i = 0
            for line in amatfile.readlines():
                if i < 2 :
                    i += 1
                    continue
                else :
                  line_list = line.strip('\#size:').split()
                  #print(line_list)
                  csvwriter.writerow(line_list)
```
