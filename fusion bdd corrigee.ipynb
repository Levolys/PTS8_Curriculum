# Script

```
def fusion_csv(l,newFileName) :
  
  with open(newFileName, 'wb') as csvfile:
    csvwriter = csv.writer(csvfile, dialect='excel')
    
    isFirst = True
    
    for bdd in l :
      
      with open(bdd, 'r') as f :

          ind = 0
          for line in f.readlines():
            
            if ind == 0 :
              if isFirst :
                csvwriter.writerow(line.split(","))
                isFirst = False
              ind += 1
              
            else :
              csvwriter.writerow(line.split(","))
        
```
